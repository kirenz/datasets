# Data hitters-clean

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/kirenz/datasets/master/Hitters.csv").dropna()

dummies = pd.get_dummies(df[['League', 'Division','NewLeague']])

df_ = df.drop(['League', 'Division', 'NewLeague'], axis=1).astype('float')
df_new = pd.concat([df_, dummies[['League_N', 'Division_W', 'NewLeague_N']]], axis=1)

df_new.to_csv("~/downloads/hitters-clean.csv", index=False)
```
